```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return np.nan
```

```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df[feature] = df[feature].apply(literal_eval)
```

```python
df['director'] = df['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
  df[feature] = df[feature].apply(get_list)
```

```python
productions = ['production_companies', 'production_countries']
df_cleaned = df.copy()

for prod in productions:
  df_cleaned[prod] = df[prod].apply(literal_eval)
  df_cleaned[prod] = df_cleaned[prod].apply(get_list)
```

```python
def get_info(data, col, is_str=True):
  if is_str:
    data[f'count_{col}'] = data[col].apply(lambda x: len(str(x).split()))
  else:
    data[f'count_{col}'] = data[col].apply(lambda x: len(x))
  lengths = list(data[f'count_{col}'])
  print(f"Number of {col}:",len(lengths),
        "\nAverage word count", np.average(lengths),
        "\nMinimum word count", min(lengths),
        "\nMaximum word count", max(lengths))
```

- v - the number of votes for the movie;
- m - the minimum votes required to be listed in the chart;
- R - the average rating of the movie;
- C - the mean vote across the whole report

```python
m = df_cleaned_types['vote_count'].quantile(0.96)
C = df_cleaned_types['vote_average'].mean()
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return round((v/(v+m) * R) + (m/(m+v) * C), 3)
```

```python
df_cleaned_types['weight_rating'] = df_cleaned_types.apply(weighted_rating, axis=1)
```
